### Data Analysis

In this notebook, I read in some of the data points. Specifically, I display the audio samples through an interactive widget, show waveform plots and Mel Spectrograms for three samples from each class (positive, negative). 

```python
import os
from glob import glob
import librosa
import IPython
from IPython.display import Audio
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
```

## Display audio samples 

```python
def plot_wave(time, y):
    fig, ax = plt.subplots()
    ax.plot(time, y)
    ax.set(xlabel='Time (seconds)', ylabel='Amplitude')
    plt.show()
```

```python
def plot_mel(y, sr):
    # generate the spectrogram
    time = np.arange(0, len(y)) / sr
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,
                                    fmax=8000)
    # plot 
    fig, ax = plt.subplots()
    S_dB = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_dB,x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax)
```

The following data points below were picked randomly.

```python
sample_one = "mod_data/neg/51.wav" 
y, sr = librosa.load(sample_one)
time = np.arange(0, len(y)) / sr
```

```python
plot_wave(time, y) 
plot_mel(y, sr)
```
